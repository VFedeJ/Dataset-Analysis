##### En que año se lanzaron mas Peliculas y Series ?

```python
#Vemos que la data esta en formato object, por lo que debemos convertirla en tipo de dato DateTime
data.Release_Date
```

```python
#Creamos nueva columna 'fecha' para tomar el dia de lanzamiento en formato datetime
data['fecha']=pd.to_datetime(data['Release_Date'])
```

```python
data.head()
```

```python
#A partir de la columna fecha hacemos obtenemos el año
data['year']=data['fecha'].dt.year
```

```python
data.head()
```

```python
#Agrupamos por 'year' y contamos la cantidad de shows por cada año
Años=data.groupby('year').Show_Id.count()
Años
```

```python
#Graficamos los valores en un chart de lineas
plt.figure(figsize=(10,6))
plt.plot(Años,marker="*",label="a")
plt.xlabel("Año")
plt.ylabel("Peliculas y Series")
plt.title("Cantidad de Series y Peliculas por Año")
plt.show()
```

### Cual es la serie o pelicula con mas duracion?

```python
data.head(2)
```

```python
data[data['Category']=='TV Show'].Duration.unique()
```

```python
data[data['Category']=='Movie'].Duration.unique()
```

```python
def convertir_en_numero(x):
    tokens=x.split(' ')
    return float(tokens[0])
    
```

```python
comida=data['Duration'].apply(convertir_en_numero)
```

```python
data['Temporadas(Tv)-Minutos(Movie)']=data['Duration'].apply(convertir_en_numero)
```
