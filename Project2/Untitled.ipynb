# CARS DATASET

###### Dataset con el detalle de autos, el costo de su compra y las caracteristicas
La analizaremos utilizando Pandas

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('CarsDataset.csv')
```

```python
df.shape
```

```python
df.head()
```

### Encontremos y reemplacemos todos los valores nullos de cada columna por la media

```python
df.isnull().sum()
```

```python
#Para cada columna reemplazamos los valores nullos por su moda (existen valores categoricos)
for column in df.columns:
    df[column].fillna(df[column].mode()[0],inplace=True)
        
```

```python
#Chequeamos que no haya mas valores nulos
df.isnull().sum()
```

### Chequear cuales son los diferentes tipos de 'Make' que hay en el dataset. Chequar la ocurrencia de valores de cada uno

```python
df['Make'].value_counts()
```

### Filtrar todos los registros de los cuales el 'Origin' sea Asia o Europa

```python
#Una opcion es esta
df[(df['Origin']=='Asia')|(df['Origin']=='Europe')]
```

```python
#Otra opcion para filtrar por varios
df[df['Origin'].isin(['Asia','Europe'])]
```

### Remover valores no deseados

Remover los registros donde Weight es mayor a 400

```python
df[df['Weight']>4000].index.values
```

```python
df[df['Weight']>4000].index.values.shape[0]
```

```python
df.drop(df[df['Weight']>4000].index.values,axis=0)
```

```python
#Otra forma de hacerlo es filtrando por los que NO son mayores a 4000 con ~
df[~(df['Weight']>4000)]
```

### Aplicamos funcion a una columna

Le sumamos 3 a todos los valores de la columna 'MPG_City'

```python
df['MPG_City']=df['MPG_City'].apply(lambda x:x+3)
```
